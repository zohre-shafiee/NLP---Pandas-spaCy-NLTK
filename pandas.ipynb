```python
Math = pd.DataFrame({'std_number':[1,2,3,4,5], 'name':['Ali','Hasan', 'Kasra','Ahmad','Hamed']})
Data_mining = pd.DataFrame({'std_number':[1,2,3,4,5], 'name':['Kamran','Hasan', 'Ahmad', 'Reza', 'Milad']})
```

```python
Math
```

```python
Data_mining
```

```python
#inner merge
pd.merge(Math, Data_mining, how = 'inner', on = 'name')
```

```python
#left and right merge
pd.merge(Math, Data_mining, how = 'left', on = 'name')
```

```python
#left and right merge
pd.merge(Math, Data_mining, how = 'right', on = 'name')
```

```python
#outer merge
pd.merge(Math, Data_mining, how = 'outer', on = 'name') 
```
