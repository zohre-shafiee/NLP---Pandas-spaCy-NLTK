```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
text= "These are test sentences, and have multiple words."
lemmatizer = WordNetLemmatizer()

tokens = word_tokenize(text)

lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]

lemmatized_words
```

```python
print(lemmatizer.lemmatize("scarves"))
print(lemmatizer.lemmatize("went", pos="v"))
print(lemmatizer.lemmatize("worst", pos="a"))
```

```python
normalizer=Normalizer()

#Exemple persian sentence
sentence= "ما به مدرسه می رفتیم"
#normalize the sentence
normalized_sentence = normalizer.normalize(sentence)

#tokenize the normalizes sentence
tokens= word_tokenize(normalized_sentence)

#lemmatize each token
lematized_tokens = [lemmatizer.lemmatize(token)for token in tokens]

print(lematized_tokens)
print(' '.join(lematized_tokens))
```
